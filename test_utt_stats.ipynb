```python
import pandas as pd
```

```python
test_all_tsv = 'data/test_all_utts2.tsv'
test_df = pd.read_csv(test_all_tsv, sep='\t')
```

```python
test_df
```

```python
out_df_file = 'out/results/test_file_stats.tsv'
```

```python
seen_spkrs = ['A04', 'A05', 'A07', 'A10', 'B03', 'B08', 'B11', 'B16']
```

```python
test_df.rename(columns={'dur': 'utt_dur'}, inplace=True)
```

```python
test_df['seen_spkr'] = test_df['spkr_id'].isin(seen_spkrs)
```

```python
# counts number of tagged, non-foreign words. e.g: pause, filled pause
# divides this number over total number. returns proportion
def get_contam_amt(text: str) -> float:
    words = text.split()
    total_words = len(words)
    contam_words = 0
    for word in words:
        if '_' in word and not word.startswith('_'):
            contam_words += 1
            
    return round(float(contam_words/total_words), 3)
```

```python
test_df['contam_amt'] = test_df['text'].apply(get_contam_amt)
```

```python
test_df.to_csv(out_df_file, sep='\t', index=False)
```
