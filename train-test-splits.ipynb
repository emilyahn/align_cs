```python
# copy textgrids and wav files to MFA input folders
all_tg_dir = 'data/mfa_input/all_tg3'
all_wav_dir = 'data/all_utts'
```

```python
urum_dir = 'data/mfa_input/train/urum_train3/'
for train_urum_id in train_urum_df['utt_id']:
    train_urum_tg = f'{all_tg_dir}/{train_urum_id}.TextGrid'
    train_urum_wav = f'{all_wav_dir}/{train_urum_id}.wav'
    if os.path.exists(train_urum_tg):
        shutil.copy(train_urum_tg, urum_dir)
        shutil.copy(train_urum_wav, urum_dir)
```

```python
cs_dir = 'data/mfa_input/train/cs_train3/'
for train_urum_id in final_train_cs_df2['utt_id']:
    train_urum_tg = f'{all_tg_dir}/{train_urum_id}.TextGrid'
    train_urum_wav = f'{all_wav_dir}/{train_urum_id}.wav'
    if os.path.exists(train_urum_tg):
        shutil.copy(train_urum_tg, cs_dir)
        shutil.copy(train_urum_wav, cs_dir)
```

```python
test_dir = 'data/mfa_input/test3/'
for test_urum_id in test_df['utt_id']:
    test_urum_tg = f'{all_tg_dir}/{test_urum_id}.TextGrid'
    test_urum_wav = f'{all_wav_dir}/{test_urum_id}.wav'
    if os.path.exists(test_urum_tg):
        shutil.copy(test_urum_tg, test_dir)
        shutil.copy(test_urum_wav, test_dir)
```
